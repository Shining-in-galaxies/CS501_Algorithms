Problem1-Leetcode#912-Sort an Array-Medium

Given an array of integers nums, sort the array in ascending order and return it.

You must solve the problem without using any built-in functions in O(nlog(n)) time complexity and with the smallest space complexity possible.



Example 1:

Input: nums = [5,2,3,1]
Output: [1,2,3,5]
Explanation: After sorting the array, the positions of some numbers are not changed (for example, 2 and 3), while the positions of other numbers are changed (for example, 1 and 5).
Example 2:

Input: nums = [5,1,1,2,0,0]
Output: [0,0,1,1,2,5]
Explanation: Note that the values of nums are not necessairly unique.

```python
from typing import List

class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        self.mergeSort(nums, 0, len(nums) - 1)
        return nums

    def mergeSort(self, arr: List[int], s: int, e: int) -> List[int]:
        # Check if it's one-element array
        if e - s + 1 <= 1:
            # If it is one-element array, we just return it
            return arr

        # If it's not an one-element array:
        m = (s + e) // 2 # m is the middle one

        self.mergeSort(arr, s, m)
        self.mergeSort(arr, m + 1, e)
        self.merge(arr, s, m, e)
        return arr

    def merge(self, arr: List[int], s: int, m: int, e: int) -> None:
        # Copy the sorted left & right halves to temp arrays
        L = arr[s:m + 1]
        R = arr[m + 1:e + 1]

        i = 0 # index for L
        j = 0 # index for R
        k = s # index for arr

        # Merge the two sorted halves into the original array
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # One of the halves will have elements remaining
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
```

Problem2-Leetcode#23-Merge k- Sorted Lists-Hard

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.



Example 1:

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6
Example 2:

Input: lists = []
Output: []
Example 3:

Input: lists = [[]]
Output: []


Constraints:

k == lists.length
lists[i] is sorted in ascending order.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

import heapq  # Importing the heapq module to use the heap data structure

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        values = []  # Initialize an empty list to store the values of all nodes

        # Traverse each linked list in the input list 'lists'
        for node in lists:
            while node:
                values.append(node.val)  # Append the value of each node to 'values'
                node = node.next  # Move to the next node in the list

        # If 'values' is empty, return None (no nodes to merge)
        if not values:
            return None

        heapq.heapify(values)  # Convert 'values' list into a min-heap

        dummy = ListNode(0)  # Create a dummy node to serve as the start of the merged list
        curr = dummy  # 'curr' will point to the current node in the merged list

        # Extract the smallest elements from the heap and create new nodes in the merged list
        while values:
            val = heapq.heappop(values)  # Pop the smallest value from the heap
            curr.next = ListNode(val)  # Create a new node with the popped value and link it to 'curr'
            curr = curr.next  # Move 'curr' to the newly created node

        return dummy.next  # Return the merged list, which starts at 'dummy.next'
```

Problem3-Leetcode#215-Kth Largest Element in an Array-Medium

Given an integer array nums and an integer k, return the kth largest element in the array.

Note that it is the kth largest element in the sorted order, not the kth distinct element.

Can you solve it without sorting?



Example 1:

Input: nums = [3,2,1,5,6,4], k = 2
Output: 5
Example 2:

Input: nums = [3,2,3,1,2,4,5,5,6], k = 4
Output: 4

```python
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        # Initialize an empty list
        k_numbers_min_heap = []

        # Add first k elements to the list
        for i in range(k):
            k_numbers_min_heap.append(nums[i])

        # Convert the list into a min-heap
        heapq.heapify(k_numbers_min_heap)

        # Loop through the remaining elements in the 'nums' array
        for i in range(k, len(nums)):
            # Compare the current element with the minimum
            # element (root) of the min-heap
            if nums[i] > k_numbers_min_heap[0]:
                # Remove the smallest element
                heapq.heappop(k_numbers_min_heap)
                # Add the current element
                heapq.heappush(k_numbers_min_heap, nums[i])

        # The root of the heap has the Kth largest element
        return k_numbers_min_heap[0]
```

Problem4-Leetcode#75-Sort Colors-Medium

Given an array nums with n objects colored red, white, or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white, and blue.

We will use the integers 0, 1, and 2 to represent the color red, white, and blue, respectively.

You must solve this problem without using the library's sort function.



Example 1:

Input: nums = [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]
Example 2:

Input: nums = [2,0,1]
Output: [0,1,2]


Constraints:

n == nums.length
1 <= n <= 300
nums[i] is either 0, 1, or 2.


```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        # Initialize pointers for red, white, and blue colors
        red, white, blue = 0, 0, len(nums) - 1

        # Iterate through the list while the white pointer is within the bounds
        while white <= blue:
            if nums[white] == 0:  # Current element is red
                # Swap the current element with the element at the red pointer
                nums[white], nums[red] = nums[red], nums[white]
                red += 1  # Move the red pointer to the right
                white += 1  # Move the white pointer to the right
            elif nums[white] == 1:  # Current element is white
                white += 1  # Move the white pointer to the right
            else:  # Current element is blue
                # Swap the current element with the element at the blue pointer
                nums[white], nums[blue] = nums[blue], nums[white]
                blue -= 1  # Move the blue pointer to the left
```

Problem5:-Leetcode#147-Insertion Sort List-Medium

Given the head of a singly linked list, sort the list using insertion sort, and return the sorted list's head.

The steps of the insertion sort algorithm:

Insertion sort iterates, consuming one input element each repetition and growing a sorted output list.
At each iteration, insertion sort removes one element from the input data, finds the location it belongs within the sorted list and inserts it there.
It repeats until no input elements remain.

Input: head = [4,2,1,3]

Output: [1,2,3,4]


Input: head = [-1,5,3,4,0]

Output: [-1,0,3,4,5]

Example 3:

Input: head = []

Output: []

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Initialize back and front pointers
        back = head
        front = head.next

        stop = 0  # Initialize a counter to keep track of the number of comparisons

        # First pass to bubble the largest values towards the end
        while front:
            if back.val > front.val:  # If the current pair is out of order
                back.val, front.val = front.val, back.val  # Swap their values
            back = back.next  # Move the back pointer forward
            front = front.next  # Move the front pointer forward
            stop += 1  # Increment the counter

        # Second pass to ensure the entire list is sorted
        for _ in range(stop):
            back = head  # Reset the back pointer to the head
            front = head.next  # Reset the front pointer to the second node

            while front:
                if back.val > front.val:  # If the current pair is out of order
                    back.val, front.val = front.val, back.val  # Swap their values
                back = back.next  # Move the back pointer forward
                front = front.next  # Move the front pointer forward

        return head  # Return the sorted list
```

Problem6:-Leetcode#148-Sort List-Medium

Given the head of a linked list, return the list after sorting it in ascending order.

Input: head = [4,2,1,3]

Output: [1,2,3,4]

Input: head = [-1,5,3,4,0]

Output: [-1,0,3,4,5]

Example 3:

Input: head = []

Output: []

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Step 0: Check if the head is None (empty list), return head if true
        if not head:
            return head

        node = head  # Initialize a pointer to traverse the list
        sorted_list = []  # Initialize a list to store the node values

        # Step 1: Traverse the linked list and store the values in sorted_list
        while node:
            sorted_list.append(node.val)
            node = node.next

        # Step 2: Sort the list of values
        sorted_list.sort()

        # Step 3: Traverse the linked list again and update the node values with sorted values
        node = head
        for val in sorted_list:
            node.val = val
            node = node.next

        return head  # Return the head of the sorted linked list
```
