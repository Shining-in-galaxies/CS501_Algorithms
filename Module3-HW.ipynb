Question 1: LeetCode Questino #155. Easy. "Min Stack"

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the MinStack class:

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
You must implement a solution with O(1) time complexity for each function.


```python
class MinStack:
    def __init__(self):
        # Initialize two stacks: one for regular values and one for tracking minimum values
        self.stack = []
        self.minStack = []

    def push(self, val: int) -> None:
        # Push the value onto the main stack
        self.stack.append(val)
        # Push the minimum value onto the minStack
        # If minStack is not empty, compare the new value with the current minimum and push the smaller one
        if self.minStack:
            val = min(self.minStack[-1], val)
        self.minStack.append(val)

    def pop(self) -> None:
        # Remove the top value from both the main stack and the minStack
        self.stack.pop()
        self.minStack.pop()

    def top(self) -> int:
        # Return the top value of the main stack
        return self.stack[-1]

    def getMin(self) -> int:
        # Return the current minimum value from the minStack
        return self.minStack[-1]
```

Question 2: LeetCode Questino #1472. Medium. "Design Browser History"
You have a browser of one tab where you start on the homepage and you can visit another url, get back in the history number of steps or move forward in the history number of steps.

Implement the BrowserHistory class:

BrowserHistory(string homepage) Initializes the object with the homepage of the browser.
void visit(string url) Visits url from the current page. It clears up all the forward history.
string back(int steps) Move steps back in history. If you can only return x steps in the history and steps > x, you will return only x steps. Return the current url after moving back in history at most steps.
string forward(int steps) Move steps forward in history. If you can only forward x steps in the history and steps > x, you will forward only x steps. Return the current url after forwarding in history at most steps.

```python
class BrowserHistory:
    def __init__(self, homepage: str):
        # Initialize the browser history with the homepage
        self.arr = [homepage]
        self.end = 0  # Track the end of the current history
        self.currWebsite = 0  # Track the current website index

    def visit(self, url: str) -> None:
        # When visiting a new URL, update the history
        if self.currWebsite == self.end: 
            # If the current website is the last in history
            if self.end == len(self.arr) - 1:
                # If the array is full, append the new URL
                self.arr.append(url)
            else:
                # Otherwise, overwrite the next position
                self.arr[self.end + 1] = url
            self.end += 1
            self.currWebsite = self.end 
        else:
            # If we are not at the end of the history, overwrite the next position
            self.arr[self.currWebsite + 1] = url
            self.currWebsite += 1
            self.end = self.currWebsite

    def back(self, steps: int) -> str:
        # Move back in the history by the specified number of steps
        if self.currWebsite - steps >= 0:
            self.currWebsite -= steps
            return self.arr[self.currWebsite]
        else:
            # If steps exceed the start of the history, go to the first page
            self.currWebsite = 0
            return self.arr[0]

    def forward(self, steps: int) -> str:
        # Move forward in the history by the specified number of steps
        if self.currWebsite + steps <= self.end:
            self.currWebsite += steps
            return self.arr[self.currWebsite]
        else:
            # If steps exceed the end of the history, go to the last visited page
            self.currWebsite = self.end
            return self.arr[self.end]
```
